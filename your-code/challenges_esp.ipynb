#### Ahora intenta agregar los siguientes elementos a `tup`.

¬øPuedes hacerlo? Explica.

```
"r", "o", "n", "h", "a", "c", "k',
```

````python
nuevos_elementos = ("r", "o", "n", "h", "a", "c", "k",)
print(nuevos_elementos) 
tup = tup + nuevos_elementos
print(tup)

# Las tuplas no se pueden modificar. Se crea un nueva tupla que contiene los elementos de la tupla original y los nuevos elementos.
# Puedo mantener el mismo nombre porque tup pasa a hacer referencia a la nueva tupla.
````

Tambi√©n imprime `tup1` y `tup2`.

````python
tup1 = tup[:4]
tup2 = tup[-4:]

print("tup1:", tup1)
print("tup2:", tup2)
````

#### Suma `tup1` y `tup2` en `tup3` usando el operador `+`.

Luego imprime `tup3` y verifica si `tup3` es igual a `tup`.

````python
tup3 = tup1 + tup2
print("tup3:", tup3)

print("tup3 es igual a tup?", tup3 == tup)
````

#### Cuenta el n√∫mero de elementos en `tup1` y `tup2`. Luego suma los dos conteos y verifica si la suma es igual al n√∫mero de elementos en `tup3`.

````python
recuento_tup1 =  len(tup1)
print("Elementos en tup1:", recuento_tup1)
recuento_tup2 = len(tup2)
print("Elementos en tup2:", recuento_tup2)

recuento_total = recuento_tup1 + recuento_tup2
print("La suma de elementos en las dos tuplas es:", recuento_total)

verifica_elementos = recuento_total == len(tup3)
print("La suma de los recuentos es igual al n√∫mero de elementos en tup3?", verifica_elementos)
````
